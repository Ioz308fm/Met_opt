```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#аналитическое решение системы ДУ
def anfy(x):
    return -np.sin(x)
def anfz(x):
    return 0+x*0
```

```python
def RK4(N,a,b,y0): #при условии x0=a
    h=(b-a)/N #возьмем равномерную сетку
    x=np.arange(a,b+h,h)
    y=[[None for j in range(len(y0))] for i in range(N+1)] # двумерный массив итоговых значений

    y[0]=y0
    for n in range (0, N): #y[n+1]=y_n+h*f(x_n+h/2,y_n+h/2 * f(x_n,y_n))
        k1=f(x[n], y[n])
        k2=f(x[n]+h/2, y[n]+k1*h/2)
        k3=f(x[n]+h/2, y[n]+k2*h/2)
        k4=f(x[n]+h, y[n]+k3*h)
        dy=h*(k1+2*k2+2*k3+k4)/6
        y[n+1]=y[n]+dy
    return y


def RK2(N,a,b,y0): #при условии x0=a
    h=(b-a)/N #возьмем равномерную сетку
    x=np.arange(a,b+h,h)
    y=[[None for j in range(len(y0))] for i in range(N+1)] # двумерный массив итоговых значений

    y[0]=y0
    for n in range (0, N): #y[n+1]=y_n+h*f(x_n+h/2,y_n+h/2 * f(x_n,y_n))
        t1=f(x[n], y[n])
        t1*=h/2
        y[n+1]=y[n]+h*f(x[n]+h/2, y[n]+t1)
    return y
def f(x,y):
    F=np.array([0.,0.])
    F[0]=y[1]-np.cos(x) # y'= z-cosx
    F[1]=y[0]+np.sin(x) # z'= y+sinx
    return F
N=1
a=0
b=0.5
y0=np.array([0.,0.])
res=RK2(N,a,b,y0)
res2=RK4(N,a,b,y0)

#построение графиков
plotresy=[None]*(N+1)
plotresz=[None]*(N+1)

plotres4y=[None]*(N+1)
plotres4z=[None]*(N+1)
for n in range(0,N+1):
    plotresy[n]=res[n][0]
    plotresz[n]=res[n][1]
    plotres4y[n]=res2[n][0]
    plotres4z[n]=res2[n][1]
    

x = np.linspace(a, b, N+1)
y1 = anfy(x)
y2 = anfz(x)
fig, ax = plt.subplots()
ax.plot(x, plotres4y, label='RK4 y(x)')#RK численное решение методом Рунге-Кутта 2 порядка
ax.plot(x, plotres4z, label='RK4 z(x)')
ax.set_xlim(a, b)
ax.set_ylim(-1, 1)
plt.legend()
plt.show()

fig, ax = plt.subplots()
ax.plot(x, plotresy, label='RK2 y(x)')#RK численное решение методом Рунге-Кутта 2 порядка
ax.plot(x, plotresz, label='RK2 z(x)')
ax.set_xlim(a, b)
ax.set_ylim(-1, 1)
plt.legend()
plt.show()

fig, ax = plt.subplots()
ax.plot(x, y1, label='analytic y(x)')#аналитическое решение
ax.plot(x, y2, label='analytic z(x)')
ax.set_xlim(a, b)
ax.set_ylim(-1, 1)
plt.legend()
plt.show()
```
